```python
import json
import numpy as np
import pandas as pd
```

```python
data = {"Name":"Sandy","Occupation":"Jr.Data Scientist","Location":"Hyderabad"}
```

```python
with open("flat.json","w") as f:
    json.dump(data,f,indent=4)
```

```python
df = pd.read_json("flat.json",typ = "series")
```

```python
df
```

```python
# nested json
data1 = [
    {
        "Name": "Sandy",
        "Occupation": "Junior Data Scientist",
        "Location": {
            "City": "Hyderabad",
            "State": "Telangana"
        },
        "Skills": ["Python", "SQL", "Power BI"]
    },
    {
        "Name": "Prats",
        "Occupation": "Data Analyst",
        "Location": {
            "City": "Bengaluru",
            "State": "Karnataka"
        },
        "Skills": ["Excel", "Tableau", "SQL"]
    }
]
```

```python
with open("nested.json","w") as f:
    json.dump(data1,f,indent=4)
```

```python
df = pd.read_json("nested.json",typ = "series")
```

```python
df
```

```python
# converting nested json to flat json 
from pandas import json_normalize
```

```python
df = json_normalize(data1)
```

```python
df
```

```python
df = df.rename(columns={"location.City": "City"})
```

```python
df
```

```python
df.rename(columns={"location.City": "City", "location.State": "State"}, inplace=True)
```
